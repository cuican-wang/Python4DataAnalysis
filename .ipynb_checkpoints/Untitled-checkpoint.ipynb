```python
import pandas as pd
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns
import datetime
```

```python
%matplotlib inline
```

```python
data = pd.read_csv('C:/Users/Wang/Desktop/DataAnalysisPython/data/stock_data/data/stocks/AAIT.csv')
data.head()
```

```python
new_df = data[['Date','High']]
```

```python
new_df.loc[:,'Date'] = new_df['Date'].apply(lambda x: datetime.datetime.strptime(x,'%Y-%m-%d').strftime('%Y'))
```

```python
new_df.head()
```

```python
plt.subplots(figsize=(20,8))

sns.set(style="ticks")
sns.set(style="whitegrid")
ax = sns.boxplot(x="Date",y="High",data=new_df,color="#8f99fb",)
ax.set_xticklabels(labels="Date",rotation=30)
#ax.set_yticklabels(labels="High",rotation=30)

ax.xaxis.get_label().set_fontsize(30)
ax.yaxis.get_label().set_fontsize(30)

ax.set_title("Distribution of stocks of AAIT.csv over years")
ax.title.set_fontsize(36)

ax.tick_params(axis="x",which='major',labelsize=20)
ax.tick_params(axis="y",which='major',labelsize=20)

sns.stripplot(x="Date", y="High", data=new_df,jitter=True, size=3, color=".3", linewidth=0)
#ax.set_xscale("log")
sns.despine(trim=True)
new_df("High").groupby("Date").mean()
plt.ylim((0.8*new_df))
```

```python
g = sns.FacetGrid(new_df, col="High", size=1, aspect=.4)
(g.map(sns.boxplot, "Date", "High").despine(left=True).add_legend(title="Distribution of stocks of AAIT.csv over years"))  
```

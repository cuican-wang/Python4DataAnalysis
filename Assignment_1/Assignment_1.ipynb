```python
import os
import json
import re

path_to_json = "C:/Users/Wang/Desktop/DataAnalysisPython/Lecture-03/jsons/"
files = [pos_json for pos_json in os.listdir(path_to_json) if pos_json.endswith('.json')]

d={}
for item in files:
    item=item.replace('.json','')
    for txt in open (path_to_json+item+'.json','r'):
        txt=re.sub(r"(,?)(\w+?)\s*?:", r"\1'\2':", txt)
        txt=txt.replace("'","\"")
        d[item]=json.loads(txt)
subd={}
for item in files:
    first=0
    third=0
    dd={"first":0,"third":0}
    item=item.replace('.json','')
    if "retargetingSegments" in d[item].keys():
        for y in d[item]["retargetingSegments"]:
            for(key,value) in y.items():
                if key=="segmentUid" and value.startswith("0"):
                    first=first+1
                if key=="segmentUid" and value.startswith("1"):
                    third=third+1
    dd["first"]=first
    dd["third"]=third
    subd[item]=dd
print(len(subd))
print(subd)
```
